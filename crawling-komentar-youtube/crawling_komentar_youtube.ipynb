## AMBIL SEMUA KOMENTAR DAN SAVE DALAM CSV

```python
from googleapiclient.discovery import build
import pandas as pd


# video id
# https://youtu.be/xnZTCdUSN8k?si=UyVkMXNOiRwq5tY6

# ðŸ“Œ Video ID yang benar adalah: xnZTCdUSN8k

# Ganti dengan API key kamu
API_KEY = 'MASUKAN API KAMU'  # <-- Ganti ini
VIDEO_ID = 'xnZTCdUSN8k'  # <-- Ganti ini (misal: 'dQw4w9WgXcQ')
```

```python
# Inisialisasi YouTube API
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Ambil komentar dari video
def get_comments(video_id):
    comments = []
    next_page_token = None

    while True:
        response = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            maxResults=100,
            textFormat='plainText',
            pageToken=next_page_token
        ).execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append({
                'author': comment['authorDisplayName'],
                'comment': comment['textDisplay'],
                'likeCount': comment['likeCount'],
                'publishedAt': comment['publishedAt']
            })

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return comments

# Jalankan dan simpan ke CSV
data = get_comments(VIDEO_ID)
df = pd.DataFrame(data)
df.to_csv('youtube_comments.csv', index=False, encoding='utf-8')
print(f"Berhasil menyimpan {len(df)} komentar ke file youtube_comments.csv")
```

```python
from googleapiclient.discovery import build
import pandas as pd

# Ganti dengan API key kamu
API_KEY = 'AIzaSyD8qtuZ_eM7lx2j_rPVs-lK-2aDtTuE4ZA'  # <-- Ganti ini
VIDEO_ID = 'xnZTCdUSN8k'  # <-- Ganti ini (misal: 'dQw4w9WgXcQ')

youtube = build('youtube', 'v3', developerKey=API_KEY)

def get_comments_and_replies(video_id):
    rows = []
    next_page_token = None

    while True:
        response = youtube.commentThreads().list(
            part='snippet,replies',
            videoId=video_id,
            maxResults=100,
            textFormat='plainText',
            pageToken=next_page_token
        ).execute()

        for item in response['items']:
            top = item['snippet']['topLevelComment']['snippet']
            top_author = top['authorDisplayName']
            rows.append({
                'comment': top['textDisplay'],
                'author': top_author,
                'reply_to_author': '',
                'likeCount': top['likeCount'],
                'publishedAt': top['publishedAt']
            })

            # Tambahkan tiap reply sebagai baris baru
            if 'replies' in item:
                for reply in item['replies']['comments']:
                    r = reply['snippet']
                    rows.append({
                        'comment': r['textDisplay'],
                        'author': r['authorDisplayName'],
                        'reply_to_author': top_author,
                        'likeCount': r['likeCount'],
                        'publishedAt': r['publishedAt']
                    })

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return rows

# Jalankan
data = get_comments_and_replies(VIDEO_ID)
df = pd.DataFrame(data)

# Simpan ke Excel
df.to_excel('youtube_comments_split_rows.xlsx', index=False)

# Kalau di Colab, unduh filenya
from google.colab import files
files.download('youtube_comments_split_rows.xlsx')
```
